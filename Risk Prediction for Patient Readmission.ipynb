```python
def categorize_diag(code):
        code = str(code)
        if code.startswith('V') or code.startswith('E'):
            return 'Other/Special'
        try:
            val = float(code)
            if 390 <= val <= 459:
                return 'Circulatory'
            elif 460 <= val <= 519:
                return 'Respiratory'
            else:
                return 'Other'
        except ValueError:
            return 'Unknown'

df['diag_1'] = df['diag_1'].apply(categorize_diag)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df['readmitted'].unique()
df['readmitted'] = df['readmitted'].map({'<30': 1, '>30': 0, 'NO': 0})
```

```python
df['readmitted'].value_counts()
```

```python
X = df.drop('readmitted',axis = 1)
y = df['readmitted']
```

```python
from imblearn.combine import SMOTETomek
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=15)
oe = OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1)
```

```python
categorical_columns = ["race","gender","payer_code","medical_specialty","diag_1","diag_2","diag_3","metformin","repaglinide","nateglinide","chlorpropamide","glimepiride","acetohexamide","glipizide","glyburide","tolbutamide","pioglitazone","rosiglitazone","acarbose","miglitol","troglitazone","tolazamide","examide","citoglipton","insulin","glyburide-metformin","glipizide-metformin","glimepiride-pioglitazone","metformin-rosiglitazone","metformin-pioglitazone","change","diabetesMed"]

X_train[categorical_columns] = oe.fit_transform(X_train[categorical_columns])
X_test[categorical_columns] = oe.transform(X_test[categorical_columns])
```

```python
smt = SMOTETomek()
X_train_smt,y_train_smt = smt.fit_resample(X_train,y_train)
```

```python
import xgboost
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
```

```python
xgb = XGBClassifier()
param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [3, 5, 7],
    "learning_rate": [0.05, 0.1],
}
```

```python
grid = GridSearchCV(estimator=xgb,param_grid=param_grid,cv=3,scoring = 'f1',verbose = 2,n_jobs=-1)
```

```python
grid.fit(X_train_smt,y_train_smt)
```

```python
grid.best_params_
```

```python
y_prob = grid.predict_proba(X_test)[:,1]
```

```python
y_pred = (y_prob >  0.2).astype(int)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score
print("Accuracy Score: ", accuracy_score(y_test,y_pred))
print("Confusion Matrix: \n", confusion_matrix(y_test,y_pred))
print( classification_report(y_test,y_pred))
print(roc_auc_score(y_test,y_pred))
```

```python
import shap

explainer = shap.Explainer(grid.best_estimator_,X_train_smt)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values, X_test)
```
