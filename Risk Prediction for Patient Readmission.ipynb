```python
def categorize_diag(code):
        code = str(code)
        if code.startswith('V') or code.startswith('E'):
            return 'Other/Special'
        try:
            val = float(code)
            if 390 <= val <= 459:
                return 'Circulatory'
            elif 460 <= val <= 519:
                return 'Respiratory'
            else:
                return 'Other'
        except ValueError:
            return 'Unknown'

df['diag_1'] = df['diag_1'].apply(categorize_diag)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df['readmitted'].unique()
df['readmitted'] = df['readmitted'].map({'<30': 1, '>30': 0, 'NO': 0})
```

```python
X = df.drop('readmitted',axis = 1)
y = df['readmitted']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 15)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in ["race","gender","payer_code","medical_specialty","diag_1","diag_2","diag_3","metformin","repaglinide","nateglinide","chlorpropamide","glimepiride","acetohexamide","glipizide","glyburide","tolbutamide","pioglitazone","rosiglitazone","acarbose","miglitol","troglitazone","tolazamide","examide","citoglipton","insulin","glyburide-metformin","glipizide-metformin","glimepiride-pioglitazone","metformin-rosiglitazone","metformin-pioglitazone","change","diabetesMed"]:
    le = LabelEncoder()
    X_train[col] = X_train[col].fillna("Unknown").astype(str)
    X_test[col] = X_test[col].fillna("Unknown").astype(str)
    all_values = pd.concat([X_train[col], X_test[col]])
    le.fit(all_values)
    X_train[col] = le.transform(X_train[col])
    X_test[col] = le.transform(X_test[col])    
```
