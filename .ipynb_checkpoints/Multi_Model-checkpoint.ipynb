```python
import torch
from torch.utils.data import Dataset# Oz datasetimi yaratmaq ucun istifade edilir
from PIL import Image# Sekilleri acmaq ucun istifade edilir
import os # Fayl yollari ile islemek ucun istifade edilir
import pandas as pd
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")#Burda texti integere cevirmek ucun istifade edirem
from sklearn.preprocessing import LabelEncoder # Burda SubCategoriyinin LabelEncoding edilmis hissesini Label adli sutunu menimsedirem
le = LabelEncoder()
df['label'] = le.fit_transform(df['masterCategory'])
counts = df['label'].value_counts()
df = df[df['label'].isin(counts[counts >= 100].index)]
df['label'] = df['label'].astype('category').cat.codes


image_dir = os.path.join(download_path, 'images')#Sekillerin oldugu directory-in pathini alir

class ProductDataset(Dataset):
    def __init__(self, df, tokenizer, transform=None):#dataseti yaratmaq ucun lazimli parametrleri elave edirem
        self.df = df.reset_index(drop=True)#indexleri sifirlayir
        self.tokenizer = tokenizer
        self.transform = transform

    def __len__(self):
        return len(self.df)# nece sample oldugunu gostermek ucun veririk

    def load_image(self, img_id):# sekili diskden oxumaq ucun funksiya veririk
        path = os.path.join(image_dir, f"{img_id}.jpg") #image pathleri duzeldir

        if not os.path.exists(path):
            return None

        try:
            return Image.open(path).convert("RGB")
        except:
            return None

    def __getitem__(self, idx):
        row = self.df.iloc[idx]#bu hissede sekli goturur eger sekil uygun deyilse dummy image yaradir


        image = self.load_image(row['image_id'])


        if image is None:
            image = Image.new("RGB", (224, 224))  # dummy image

        if self.transform:
            image = self.transform(image)


        text = str(row.get('productDisplayName', ""))#textleri yaradir

        text_enc = self.tokenizer(
            text, #text  rəqəmlərə çevrilir
            padding='max_length',#padding  hamısını eyni uzunluq edir
            truncation=True,#truncation  uzun texti kəsir
            max_length=64,#max_length=64 → maksimum 64 token
            return_tensors='pt'
        )


        label = torch.tensor(row['label'], dtype=torch.long)# burda category integer olur ve tensoR formatina salinir

        return {
            "image": image,#şəkil tensor
            "input_ids": text_enc['input_ids'].squeeze(0),#text rəqəm formatı
            "attention_mask": text_enc['attention_mask'].squeeze(0),#hansı tokenlər realdır
            "label": label # burda da targetler gosterilir
        }
```

```python
from torchvision import transforms
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])# burda sekli uygun olcuye salir ve tensor edir
```

```python
from torch.utils.data import DataLoader
dataset = ProductDataset(df, tokenizer, transform)# burda raw data
loader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)
```

```python
import torch
import torch.nn as nn
from transformers import DistilBertModel
from torchvision import models

class MultiModalModel(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        #TEXT MODEL (DistilBERT)
        self.text_model = DistilBertModel.from_pretrained("distilbert-base-uncased")

        #IMAGE MODEL (ResNet18)
        self.image_model = models.resnet18(pretrained=True)
        self.image_model.fc = nn.Identity()  # son layer silinir

        # FINAL CLASSIFIER
        self.classifier = nn.Sequential(
            nn.Linear(768 + 512, 256),  # 768 (text) + 512 (image)
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, input_ids, attention_mask, image):

        #  TEXT çıxışı
        text_output = self.text_model(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

        text_feat = text_output.last_hidden_state[:, 0, :]
        #(batch, 768) → CLS token

        #  IMAGE çıxışı
        image_feat = self.image_model(image)
        # (batch, 512)

        #  CONCAT
        combined = torch.cat((text_feat, image_feat), dim=1)

        # CLASSIFIER
        output = self.classifier(combined)

        return output
```

```python
import torch
import torch.nn as nn
import numpy as np
from tqdm import tqdm
from sklearn.utils.class_weight import compute_class_weight


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = MultiModalModel(num_classes=len(df['label'].unique()))
model = model.to(device)

# CLASS WEIGHTS (ƏSAS FIX)
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(df['label']),
    y=df['label']
)

weights = torch.tensor(class_weights, dtype=torch.float).to(device)

#loss
criterion = nn.CrossEntropyLoss(weight=weights)

# optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

# training
for epoch in range(5):
    model.train()
    total_loss = 0

    for i, batch in enumerate(tqdm(loader)):
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        images = batch['image'].to(device)
        labels = batch['label'].to(device)

        outputs = model(input_ids, attention_mask, images)

        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()



    print(f"Epoch {epoch} Avg Loss: {total_loss/len(loader):.4f}")


```

```python
model.eval()#Modeli test/evaluation rejiminə keçirir yeni data artiq oyrenmir sadece proqnoz verir
```

```python
all_preds = []
all_labels = []

with torch.no_grad():# burda gradient hesablanmasin deye veririk bu surete tesir edir
    for batch in loader:  #datani batch seklinde yukleyir
        input_ids = batch['input_ids'].to(device)#token-lar (DistilBERT üçün)
        attention_mask = batch['attention_mask'].to(device)#hansi tokenin real oldugunu deyir
        images = batch['image'].to(device)#CNN hissesi ucun sekiller
        labels = batch['label'].to(device)# bu da dogru cavablar

        outputs = model(input_ids, attention_mask, images)#burda datani text+image seklinde qebul edir ve cixis verir

        preds = torch.argmax(outputs, dim=1)#bu da en boyuk score-u goturur

        all_preds.extend(preds.cpu().numpy())#Gpu da olan datani cpu getirir ve numpy array seklinde geri qaytarir
        all_labels.extend(labels.cpu().numpy())# real cavablari saxlayir
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

acc = accuracy_score(all_labels, all_preds)
print("Accuracy:", acc)
print("Classification Report: \n",classification_report(all_labels, all_preds))
print('Confusion Matrix: \n',confusion_matrix(all_labels, all_preds))# Burda da metricleri predict edirik
```
